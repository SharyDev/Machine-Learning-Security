```python
######## Perform encrypted inference on the first layer ###########
###### Using Teanseal #####

import tenseal as ts
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


df = pd.read_csv('cpi-dataset.csv')

###### Features and target variable #####
X = df.drop(columns=['All-items', 'Years']).values  # Features
y = df['All-items'].values.reshape(-1, 1)  # Target

###### MinMax Scaling #####
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

###### Neural network model #####
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(units=8, activation='relu'),
    tf.keras.layers.Dense(units=1)
])

###### Compile and train the model #####
model.compile(optimizer='adam', loss='mse')
history = model.fit(X_train, y_train, epochs=100, batch_size=4, verbose=0)

###### Extract weights and biases from the first layer #####
weights_layer1 = model.layers[0].get_weights()[0]  # Weights
biases_layer1 = model.layers[0].get_weights()[1]   # Biases


########################### Homomorphic Encryption  ############################

###### Initialize TenSEAL context for CKKS scheme #####
context = ts.context(ts.SCHEME_TYPE.CKKS, poly_modulus_degree=8192, coeff_mod_bit_sizes=[60, 40, 40, 60])
context.generate_galois_keys()
context.global_scale = 2**30

###### Perform encrypted inference on the first layer #####
encrypted_predictions = []
for i in range(len(X_test)):
    ###### Encrypt each test sample #####
    encrypted_input = ts.ckks_vector(context, X_test[i].tolist())

    encrypted_result = []
    for neuron_idx in range(weights_layer1.shape[1]):
        neuron_weights = weights_layer1[:, neuron_idx]

        ###### Convert weights to pure Python floats #####
        neuron_weights_list = [float(w) for w in neuron_weights]

        neuron_bias = float(biases_layer1[neuron_idx])  # Ensure bias is a float

        ###### Perform dot product between encrypted input and neuron weights #####
        encrypted_output = encrypted_input.dot(neuron_weights_list) + neuron_bias
        encrypted_result.append(encrypted_output)

    ###### Decrypt the result from the first layer #####
    decrypted_result = [output.decrypt()[0] for output in encrypted_result]
    encrypted_predictions.append(decrypted_result)




###### Convert encrypted predictions to NumPy array #####
encrypted_predictions = np.array(encrypted_predictions)

# Since we only applied the first layer, apply ReLU manually and pass through rest of model
encrypted_predictions_relu = np.maximum(encrypted_predictions, 0)

###### Pass through second and third layers (unencrypted for demonstration) #####
second_layer_weights = model.layers[1].get_weights()[0]
second_layer_biases = model.layers[1].get_weights()[1]
third_layer_weights = model.layers[2].get_weights()[0]
third_layer_biases = model.layers[2].get_weights()[1]

###### Pass through second layer #####
second_layer_output = np.dot(encrypted_predictions_relu, second_layer_weights) + second_layer_biases
second_layer_output_relu = np.maximum(second_layer_output, 0)  # ReLU activation

###### Pass through final layer #####
final_predictions_scaled = np.dot(second_layer_output_relu, third_layer_weights) + third_layer_biases

# Denormalize predictions
final_predictions = scaler_y.inverse_transform(final_predictions_scaled)
actuals = scaler_y.inverse_transform(y_test)


plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()


print("Predictions (First 5):", final_predictions[:5].flatten())
print("Actual Values (First 5):", actuals[:5].flatten())
```

```python


import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from phe import paillier
import numpy as np


df = pd.read_csv('./cpi-dataset.csv')

# Features and target variable
X = df[['Food', 'Shelter', 'Transportation']].values  # Select some features
y = df['All-items'].values.reshape(-1, 1)  # Target

###### MinMax Scaling #####
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)


###### Generate Paillier keys #####
public_key, private_key = paillier.generate_paillier_keypair()

###### Encrypt the target variable (y_train) #####
encrypted_y_train = [public_key.encrypt(value[0]) for value in y_train]

###### Calculate coefficients using encrypted y_train and unencrypted X_train #####
n_samples, n_features = X_train.shape
coefficients = []

for i in range(n_features):
    ###### Compute numerator sum #####
    numerator = sum(X_train[j][i] * encrypted_y_train[j] for j in range(n_samples))

    ##### Compute denominator #####
    denominator = sum(X_train[:, i] ** 2)

    ###### Calculate encrypted coefficient #####
    encrypted_coef = numerator * (1 / denominator)

    coefficients.append(encrypted_coef)

###### Decrypt coefficients #####
decrypted_coefficients = [private_key.decrypt(coef) for coef in coefficients]

###### Calculate intercept #####
encrypted_intercept = sum(encrypted_y_train) * (1 / n_samples) - sum(
    decrypted_coefficients[i] * np.mean(X_train[:, i]) for i in range(n_features)
)
decrypted_intercept = private_key.decrypt(encrypted_intercept)

##### final model #####
print("Linear Regression Model Coefficients (Decrypted):")
for i, coef in enumerate(decrypted_coefficients):
    print(f"Feature {i + 1}: {coef}")
print(f"Intercept: {decrypted_intercept}")


###### Predict on unencrypted test data using decrypted coefficients #####
predictions = np.dot(X_test, decrypted_coefficients) + decrypted_intercept

###### Rescale predictions to original scale #####
final_predictions = scaler_y.inverse_transform(predictions.reshape(-1, 1))


print("Predictions on Test Data (Decrypted and Rescaled):\n", final_predictions)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error


#### Simple Model #####
unencrypted_model = LinearRegression()
unencrypted_model.fit(X_train, y_train)

###### Test Data ########
unencrypted_predictions = unencrypted_model.predict(X_test)

######## Rescale predictions ############
unencrypted_predictions_rescaled = scaler_y.inverse_transform(unencrypted_predictions)

##### Rescale actual y_test values #########
y_test_rescaled = scaler_y.inverse_transform(y_test)

###### MSE and MAE for Unencrypted Model #####
mse_unencrypted = mean_squared_error(y_test_rescaled, unencrypted_predictions_rescaled)
mae_unencrypted = mean_absolute_error(y_test_rescaled, unencrypted_predictions_rescaled)

print(f"Unencrypted Model - Mean Squared Error (MSE): {mse_unencrypted}")
print(f"Unencrypted Model - Mean Absolute Error (MAE): {mae_unencrypted}")



######  MSE and MAE for Encrypted Model  #####
mse_encrypted = mean_squared_error(y_test_rescaled, final_predictions)
mae_encrypted = mean_absolute_error(y_test_rescaled, final_predictions)

print(f"Encrypted Model - Mean Squared Error (MSE): {mse_encrypted}")
print(f"Encrypted Model - Mean Absolute Error (MAE): {mae_encrypted}")
```

```python
############### Homomorphic Encryption ###############

!pip install tenseal
import tenseal as ts
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

########### Synthetic dataset with 10 samples ##########
X, y = make_classification(n_samples=10, n_features=2, n_informative=2, n_redundant=0, n_classes=2, random_state=42)

########### Dataset Split ##########
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

###### Set up TenSEAL Context with deeper modulus chain #####
n = 32768  # polynomial degree for more capacity
q = [60, 40, 40, 40, 40, 40, 60]  # Deeper modulus chain for more operations
ctx = ts.context(ts.SCHEME_TYPE.CKKS, n, q)
ctx.generate_galois_keys()
ctx.global_scale = 2**20  #global scale for better control

####### Encrypt the training data ######
encrypted_X_train = [ts.ckks_vector(ctx, x.tolist()) for x in X_train]
encrypted_y_train = [ts.ckks_vector(ctx, [yi]) for yi in y_train]

####### Define EncryptedLogisticRegression ######
class EncryptedLogisticRegression:
    def __init__(self, n_features):
        self.weights = ts.ckks_vector(ctx, [0.0] * n_features) # initializes the weights of the model as a vector
        self.bias = ts.ckks_vector(ctx, [0.0]) # initializes the bias term of the model, also as an encrypted vector
        self.learning_rate = 0.001  #The learning rate controls how much the model updates the weights and bias during training.
        self.num_epochs = 3  # number of epochs

    def sigmoid(self, z):

        return 0.5 + 0.2 * z # standard sigmoid activation function

    def predict(self, x):
        result = self.weights + x  # Homomorphic Implementation
        result += self.bias
        result = self.sigmoid(result)
        return result
    #goal of gradient descent is to update the model's weights and bias to minimize
    #the error between predictions and actual values.
    def gradient_descent(self, encrypted_X, encrypted_y):
        m = len(encrypted_X)
        inv_m = 1.0 / m  # m = number of samples

        for epoch in range(self.num_epochs):
            dw = ts.ckks_vector(ctx, [0.0] * len(self.weights.decrypt()))
            db = ts.ckks_vector(ctx, [0.0])

            for i in range(m):
                y_pred = self.predict(encrypted_X[i])
                error = y_pred - encrypted_y[i]


                dw += encrypted_X[i] + error
                db += error

            # Weight and bias updates
            self.weights -= dw * self.learning_rate * inv_m
            self.bias -= db * self.learning_rate * inv_m

    def fit(self, encrypted_X_train, encrypted_y_train):
        self.gradient_descent(encrypted_X_train, encrypted_y_train)
  #### Prediction ######
    def score(self, X_test, y_test):
        predictions = []
        for x in X_test:
            encrypted_x = ts.ckks_vector(ctx, x.tolist())
            encrypted_pred = self.predict(encrypted_x)
            decrypted_pred = encrypted_pred.decrypt()[0]
            predictions.append(decrypted_pred)

        predicted_labels = [1 if p > 0.5 else 0 for p in predictions]
        return accuracy_score(y_test, predicted_labels)

####### Train the model ######
model = EncryptedLogisticRegression(n_features=X_train.shape[1])
model.fit(encrypted_X_train, encrypted_y_train)

####### Evaluate the model ######
accuracy = model.score(X_test, y_test)
print("Encrypted Logistic Regression Accuracy:", accuracy)
```
